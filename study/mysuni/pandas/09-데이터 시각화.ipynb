# Matplotlib

```python
import matplotlib.pyplot as plt
x = np.arange(10)
y = x ** 2
plt.plot(x, y, 'rv') # x와 y의 데이터 개수는 일치
plt.xticks(x)  # X축 간격
plt.show()
```

```python
# 도화지(Figure) : Cell을 벗어날 수 없음
# 그래프를 한 도화지(Figure)에 여러번 그릴 수 있음 -> X축과 Y축을 공유

# 1) Figure 설정 -> 크기 설정, subplot
plt.figure(figsize=(10, 5)) # (가로, 세로) -> 단위 : 인치

# 2) 그래프 추가 -> plot, bar, scatter, etc 
plt.plot(x, x**2, 'bo', label='y=x^2') # label : 그래프의 이름 지정 -> 범례 
plt.plot(x, x**3, 'g^', label='y=x^3')
plt.plot(x, x**4, 'y+', label='y=x^4')

# 3) 그래프 설정 
plt.xlabel('X축 이름') # X축 이름
plt.ylabel('Y축 이름') # Y축 이름 
plt.title('그래프 제목') # 그래프 제목
plt.legend() # 범례 표시 -> 자동으로 결정 loc로 설정 가능 (upper left)
plt.xticks(np.arange(-3, 13, 3)) # X축의 틱(눈금) 간격
plt.yticks(np.arange(0, 10000, 2000)) # Y축의 틱(눈금) 간격
plt.grid() # 틱을 기준으로 격자 추가

#plt.xlim( (3, 6)) # X축을 3에서 6사이 만 표현 
#plt.ylim() 

plt.savefig('image') # show() 이전에 사용 -> png 

plt.show() # 그래프 출력 
```

# Seaborn

```python
# DataFrame -> plot() or seaborn
import seaborn as sns
# sns.함수(data=df, x='컬럼명', y='컬럼명')
tips = sns.load_dataset('tips') # 그래프 시각화를 위한 연습용 데이터 세트 제공
tips
```

```python
sns.get_dataset_names() # 사용가능한 그래프들
```

```python
# # 1) 컬럼명으로 축 지정
# sns.boxplot(data=tips, x='total_bill')

# # 2) 값을 직접 대입
# sns.boxplot(x=tips['total_bill'])


# # X축 : 값 -> 수평
# # Y축 : 값 -> 수직

# # Boxplot -> 연속형 변수 1개, 범주형 변수 1개
# sns.boxplot(data=tips, y='total_bill', x='time') # x값에 따라서 boxplot이 그려짐

# hue : 그래프를 분할하는 역활 
sns.boxplot(data=tips, y='total_bill', x='time', hue='smoker')
```

```python
plt.figure(figsize=(8, 8))
sns.lmplot(data=tips, x='total_bill', y='tip', hue='time') # X축, Y축 모두 연속형 변수 -> 실수 -> 상관 관계
# sns.regplot
plt.show()
```

```python
# heatmap : 값을 색상으로 표현 -> pivot_table(dataframe)의 결과와 잘 어울림
plt.figure(figsize=(15, 15))
sns.heatmap(df_pv.T / 100, cmap='Blues', annot=True, fmt='.2f')
# cmap : 색상 범례 (color bar, cbar=True)
    # jet : (낮) 파-빨 (높)
    # binary : (낮) 흰-검 (높)
    # Greens : 초록색
    # Blues : 파란색
# annot : 숫자 출력 유무 
# fmt : 숫자 형식 
 # d : 정수
 # f : 실수 (6자리) -> .2f (2자리)
plt.show()
```
