```python
import pandas as pd
import numpy as np
```

# 데이터 프레임 생성
1) 100에서 200 사이의 값을 60개를 생성합니다(np.random.randint 이용)

2) 생성한 결과를 12,5 형태로 변환합니다.

3) 변환된 데이터를 이용하여 df 데이터 프레임을 생성합니다.

```python
np.random.seed(1234) # 랜덤 -> 랜덤X (값고정)
df = pd.DataFrame(np.random.randint(100,200,60).reshape(12, 5))
df
```

# 컬럼명 변경
- 컬럼명을 ['P01', 'P02', 'P03', 'P04', 'P05']로 수정하세요.

```python
# columns 속성 이용
df.columns = ['P01', 'P02', 'P03', 'P04', 'P05']
df
```

# 인덱스 변경
- 인덱스를 2020년 1월 부터 2020년 12월까지 마지막 요일로 설정하세요. 

```python
# index 생성 
# date_range 함수 이용 (주기는 월(M)으로 설정)
date = pd.date_range('20200101','20201231', freq='M')
display(date)
# index 속성 이용하여 date 변수를 index로 지정하세요.
df.index = date
df
```

```python
df.drop([인덱스, 인덱스, ...], axis=0, inplace=True) # 행 삭제
df.drop([컬럼명, 컬럼명, ...], axis=1, inplace=True) # 열 삭제
```

```python
df.loc[df.index[3]:df.index[5]]
```

- 2020년 05월 31일의 P01 컬럼 ~ P03 컬럼을 출력하세요.

```python
df.loc["2020-05-31", 'P01':'P03'] 
# df.loc["2020-05-31", 'P01', 'P02', 'P03']
# df.loc["2020-05-31", df.columns.isin(['P01', 'P02', "P03"])]
```

- 2020년 01월 31일의 P01 컬럼부터 P05 컬럼의 값을 0으로 설정하세요.

```python
# df.loc[행 인덱스, 열 인덱스] = 값
df.loc["2020-01-31", :] = 0 
df
```

- 각 행의 합을 이용하여 Total 컬럼을 생성하세요
    - total = P01 + P02 + P03 + P04 + P05

```python
df['Total'] = df['P01'] + df['P02'] + df['P03'] + df['P04'] + df['P05']
df
```

```python
# df[새로운컬럼명] = 값
df['Total'] = df.loc[:,'P01':'P05'].sum(axis=1) # 인덱스가 일치하는게 없어서 NaN
df
```

- Total 컬럼을 Total 컬럼의 총합 값으로 나누어 Total Rate 컬럼을 생성하세요.

```python
df['Total Rate'] = df['Total'] / df['Total'].sum() * 100
df.round(2)
```

- Total Rate 컬럼이 가장 낮은 3달을 출력하세요.

```python
df.sort_values(by=['Total Rate']).head(3)
```

- Total 값이 700 미만인 데이터를 출력하세요.

```python
df.loc[df['Total'] < 700, :]
```

- Total 값이 700 미만이면 Bad 700 이상이면 Good의 값을 가지는 state 컬럼을 생성하세요.

```python
df.loc[df['Total'] < 700, 'state'] = 'Bad' # 조건에 해당되지 않는 경우 NaN 들어감
df.loc[df['Total'] >= 700, 'state'] = 'Good'
df
```

- 합계 행을 추가하세요.

```python
df.index[-2]
```

```python
df.loc['합계', 'P01':'Total'] = df.loc[:df.index[-2], :].sum()
df
```

- Total 컬럼의 합계 값을 이용하여 P01부터 P05까지의 비율을 계산 후 합계 비율 행을 추가하세요.

```python
df.loc['합계 비율', :] = df.loc['합계', 'P01':'P05'] / df.loc['합계', 'Total']
df
```
