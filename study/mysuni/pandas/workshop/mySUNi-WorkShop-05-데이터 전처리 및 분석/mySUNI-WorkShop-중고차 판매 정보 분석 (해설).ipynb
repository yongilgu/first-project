```python
# 데이터 분석을 위한 패키지를 불러옵니다.
import pandas as pd                   # 데이터 분석을 위한 모듈
import matplotlib.pyplot as plt       # 데이터 시각화를 위한 기본 모듈
import seaborn as sns                 # 데이터 시각화를 위한 추가 모듈
```

```python
# vehicles.csv 파일을 읽어 df 변수에 저장하세요.
df = pd.read_csv('vehicles.csv')
df
```

```python
# Unnamed: 0 컬럼이 존재하면 제거하세요.
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
# 가격 컬럼의 단위를 만원으로 변경하세요.
df['가격'] = df['가격'] / 10000
df
```

```python
# 제조사별 판매 차량 수를 계산하세요.
df['제조사'].value_counts()
```

```python
# 유형별 판매 차량 수를 계산하세요.
df['유형'].value_counts()
```

```python
# 가격이 가장 큰 데이터 6개를 출력하세요.
df.sort_values('가격', ascending=False).head(6)
```

```python
# 주행거리가 가장 큰 데이터 6개를 출력하세요.
df.sort_values('주행거리', ascending=False).head(6)
```

```python
# 차량 모델별 평균 주행거리와 가격을 계산하세요.
result=df.groupby(['모델'])[['가격', '주행거리']].mean()
result
```

```python
# 평균 가격이 상위 10개인 차량 모델 정보를 시각화하세요.
top10=result.sort_values('가격', ascending=False).head(10)
top10.plot(kind='bar')
plt.show()
```

```python
# 주행거리가 100000보다 큰 데이터만 추출하세요.
long = df[df['주행거리'] > 100000]
long
```

```python
# 주행거리가 100000보다 큰 데이터의 차량 유형별 평균 가격을 계산하세요.
result = long.groupby('유형')['가격'].mean().sort_values(ascending=False)
result
```

```python
# 앞서 계산한 결과를 시각화하세요.
result.plot(kind='bar')
plt.show()
```

```python
# 제조사와 유형별 최소, 최대, 평균 가격을 계산하세요.
result = df.groupby(['제조사', '유형'])['가격'].agg(['min', 'max', 'mean'])
result
```

```python
# 제조사가 BMW인 유형별 평균 가격을 시각화하세요.
bmw = result.loc['bmw', :]
bmw.plot.pie(y='mean', figsize=(10, 10), autopct='%.2f%%', fontsize=15)           # 합계는 sum, 평균은 mean 
plt.show()
```

```python
# 제조사가 BMW인 유형별 최소, 최대, 평균 가격을 시각화하세요.
bmw.plot(kind='barh')
plt.show()
```

```python
# 제조사가 BMW인 데이터를 추출하여 유형별 가격의 분포와 이상치를 시각화하세요.
bmw = df[df['제조사'] == 'bmw']
plt.figure(figsize=(20, 10))
sns.boxplot(data=bmw, x='유형', y='가격')
plt.show()
```

```python
# 연식별 차량 가격의 분포와 이상치를 시각화하세요.
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, x='연식', y='가격')
plt.show()
```

```python
# 2010년도 이후 연식 데이터만 추출하여 차량 가격의 분포와 이상치를 시각화하세요.
new = df[df['연식'] >= 2010]
plt.figure(figsize=(20, 10))
sns.boxplot(data=new, x='연식', y='가격')
plt.show()
```

```python
# 2010년도 이후 연식별 주행거리의 분포와 이상치를 시각화 하세요.
new = df[df['연식'] >= 2010]
plt.figure(figsize=(20, 10))
sns.boxplot(data=new, x='연식', y='주행거리')
plt.show()
```

```python
# elantra의 연식별 감가 계산을 위해 모델이 elantra인 데이터만 추출하세요.
elantra = df[df['모델'] =='elantra']
elantra
```

```python
# 기준(2020)에서 연식을 빼  사용연도 컬럼을 생성하세요.
elantra['사용연도'] = 2020 - elantra['연식']
elantra
```

```python
# 사용연도컬럼을 기준으로 평균 금액을 계산하세요.
result=elantra.groupby(['사용연도'])['가격'].mean()
result
```

```python
# 결과를 시각화하세요.
result.plot()
```

```python
# 함수를 정의하여 모델을 변경해보세요
def plot_depreciation(model):
    temp = df[df['모델'] == model]
    temp['사용연도'] = 2020 - temp['연식']
    temp.groupby(['사용연도'])['가격'].mean().plot()
```

```python
plot_depreciation('328i')
```

```python
plot_depreciation('c-class')
```

```python
plot_depreciation('a4')
```
